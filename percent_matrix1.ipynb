```python
import itertools
```

```python
import pandas as pd
```

```python
prices = [23.88,16.09, 17.95,17.95,20.15]
```

```python
variations = itertools.permutations(prices,2)
```

```python
data = pd.DataFrame(variations, columns = ["first", "second"])
data
```

```python
data["change"] = ((data["second"]-data["first"])/data["first"] * 100).round(2)
```

```python
data
```
