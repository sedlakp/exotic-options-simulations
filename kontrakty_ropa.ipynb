```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
```

```python
may = pd.read_csv("datasets/CLK20.NYM_com.csv").set_index("Date").iloc[:,3]
may.name = "May 20"
may.index = pd.to_datetime(may.index)
jun = pd.read_csv("datasets/CLM20.NYM_com.csv").set_index("Date").iloc[:,3]
jun.name = "Jun 20"
jun.index = pd.to_datetime(jun.index)
jul = pd.read_csv("datasets/CLN20.NYM_com.csv").set_index("Date").iloc[:,3]
jul.name = "Jul 20"
jul.index = pd.to_datetime(jul.index)
aug = pd.read_csv("datasets/CLQ20.NYM_com.csv").set_index("Date").iloc[:,3]
aug.name = "Aug 20"
aug.index = pd.to_datetime(aug.index)
```

```python
import matplotlib.dates as mdates
week = mdates.WeekdayLocator(byweekday=1, interval=1, tz=None)
fmt = mdates.DateFormatter('%D')

fig, ax = plt.subplots(1, 1, figsize=(8,4.5))
ax.set_title("Po sobě následující futures kontrakty na ropu", fontsize=18)
ax.set_ylabel("$",fontsize=14)
ax.plot(may)
ax.plot(jun)
ax.plot(jul)
ax.plot(aug)
ax.legend()
ax.xaxis.set_major_locator(week)
ax.xaxis.set_major_formatter(fmt)
fig.autofmt_xdate()
plt.show()
fig.savefig('4_futures_contracts.pdf', format='pdf', bbox_inches='tight')
```
