# Asian option
### fixed strike, arithmetic average

Function is defined [here](files/simulations/asian_mean_reverting.py) (parameters are explained there as well)

Returns price, simulated paths and array of timesteps (use)

```python
price, paths, time_steps = mean_reverting_asian_fixed_strike(
    s=42,
    x=40,
    positionFlag="c",
    time_to_maturity=1,
    steps=250,
    simulations=100,
    kappa=0.0001,
    theta=45,
    v=2,
    r=0.1)
```

#### Simulated price

```python
print("Simulated option price: ",price.round(4))
```

#### Plot graph

```python
for path in paths:
    plt.plot(time_steps, path)
plt.title("Asian option (fixed strike)")
plt.xlabel("Number of steps")
plt.ylabel("Spot price")
plt.show()
```
